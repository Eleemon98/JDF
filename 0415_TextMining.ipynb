```python
# 형태소만 토큰라이징
mor_txt = okt.morphs(text)
print(mor_txt)
```

```python
def count_txt(lst):
    txt_cnt = dict()

    for i in mor_txt:
        if i not in txt_cnt.keys():
            txt_cnt[i]=1
        else:
            txt_cnt[i]+=1
    return txt_cnt
```

```python
dic_mor = count_txt(mor_txt)
```

```python
sorted(dic_mor.items(), reverse=True, key = lambda item: item[1])
```

```python
noun_txt = [t[0] for t in pos_txt if t[1]=='Noun']
noun_txt
```

```python
# 각 형태소별 빈도수
from collections import Counter

nv_cnt = Counter(noun_txt)
# 각 형태소별 빈도수 순위
nv_cnt.most_common(10)
```

```python
# 우리가 원하는 품사의 형태소만 추출
morph_lst=[]

for txt in pos_txt:
    if txt[1]=='Adjective' or txt[1]=='Noun' or txt[1]=='Adverb':
        morph_lst.append(txt[0])

print(morph_lst)
```

```python
# 2021년 신년사를 워드 클라우드 만들기
from konlpy.tag import Okt
# 형태소 분석기 선택
okt = Okt()

# 신년사 파일 열기
f = open('2021_newyear.txt', 'r', encoding='utf-8')
text = f.read()
f.close()
```

```python
# 신년사에서 명사만 추출하기
nouns = okt.nouns(text)
# 전체 단어의 개수 100개 선택
from collections import Counter
top_100 = Counter(nouns).most_common(100)
top_100
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Gulim'
wc = WordCloud(font_path='C:\Windows\Fonts\GULIM.ttc', background_color='white', colormap = 'summer')
wordcl = wc.generate_from_frequencies(dict(top_100))
plt.imshow(wordcl)
plt.axis('off')
plt.show()
```

```python
# BoW(Bag of Words)
txt = ['파이썬 차트 파이썬 머신러닝', # 문장 1
       '차트 파이썬 RR 차트', # 문장 2
       'RR 분석 시각화'] # 문장 3
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer()
vec.fit_transform(txt).toarray()
```

```python
vec.vocabulary_
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer(stop_words=['차트']) # 불용어 지정
value = vec.fit_transform(txt).toarray()
```

```python
vec.vocabulary_
```

```python
key = vec.get_feature_names_out()
```

```python
import pandas as pd

df = pd.DataFrame(value)
df.columns= key
df
```

```python
# DTM
txt = [['먹고 싶은 사과'], #doc 1
       ['먹고 싶은 바나나'], #doc 2
       ['길고 노란 바나나 바나나'], #doc 3
       ['저는 과일이 좋아요']] # doc 4
```
